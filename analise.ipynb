# Estudo dos diferentes casos para as pesquisas

1. Melhor caso - Primeiro elemento da lista
2. Pior caso - Elemento que não se encontra na lista
3. Caso intermédio - Um elemento aleatório da lista

Nos exemplos usei as lista com 100 elementos pois é mais rápido a analisar

## Listas Ordenadas

### Estudo do melhor caso, pior caso e do caso intermédio na função da pesquisa sequencial 

```python
quantidades = [1000000]
valor_max = 1000000

tempos_pesquisa_sequencial_ordenado = {"melhor_caso": [], "pior_caso": [], "caso_intermedio": []}

for quantidade in quantidades:
    ficheiro_ordenado = f'ordenado_{quantidade}.txt'
    with open(ficheiro_ordenado, 'r') as ficheiro:
        numeros = ficheiro.readlines()
        inteiros = [int(numero) for numero in numeros]
        
    item_melhor_caso = inteiros[0]

    item_pior_caso = -1

    item_caso_intermedio = inteiros[random.randint(0, len(inteiros) - 1)]

    print(f'Pesquisa Sequencial para {quantidade} ordenado (Melhor Caso):')
    dic_stats, dic_tempos = a.complexidade(p.pesquisa_sequencial, ficheiro_ordenado, [item_melhor_caso], 500)
    tempos_pesquisa_sequencial_ordenado["melhor_caso"].extend(dic_tempos[item_melhor_caso])
    print(dic_stats)

    print(f'Pesquisa Sequencial para {quantidade} ordenado (Pior Caso):')
    dic_stats, dic_tempos = a.complexidade(p.pesquisa_sequencial, ficheiro_ordenado, [item_pior_caso], 500)
    tempos_pesquisa_sequencial_ordenado["pior_caso"].extend(dic_tempos[item_pior_caso])
    print(dic_stats)

    print(f'Pesquisa Sequencial para {quantidade} ordenado (Caso Intermediário):')
    dic_stats, dic_tempos = a.complexidade(p.pesquisa_sequencial, ficheiro_ordenado, [item_caso_intermedio], 500)
    tempos_pesquisa_sequencial_ordenado["caso_intermedio"].extend(dic_tempos[item_caso_intermedio])
    print(dic_stats)
```

```python
a.complexidade_boxplot(tempos_pesquisa_sequencial_ordenado)
```
